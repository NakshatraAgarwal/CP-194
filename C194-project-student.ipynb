# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Import libraries
import pandas as pd
from matplotlib import pyplot as plt
#read the csv
df = pd.read_csv('framingham.csv')
df
```

```python
#Filter and make a new dataframe for non smokers
non_smoker = df.loc[df['currentSmoker'] == 0]
non_smoker
```

```python
#Group by age column and find average heart rate at different age

grp_nsmoker = non_smoker.groupby('age')['heartRate'].mean().reset_index()
grp_nsmoker
```

```python
#Filter and make a new dataframe for smokers
smokers = df.loc[df['currentSmoker'] == 1]
smokers
```

```python
#Group by age column and find average heart rate at different age
grp_smoker = smokers.groupby('age')['heartRate'].mean().reset_index()
grp_smoker
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers
plt.figure(figsize=(16,8))
plt.plot(grp_smoker['age'], grp_smoker['heartRate'], label="Heart Rate of People who smoke", linewidth=3)
plt.plot(grp_nsmoker['age'], grp_nsmoker['heartRate'], label="Heart Rate of People who do not smoke", linewidth = 3)
plt.xlabel("Age")
plt.ylabel('Average Heart Rate')
plt.legend()
plt.title("Comparison between Average Heart Rate of people who smoke and do not smoke")
plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv

#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
heart_disease = df.loc[df['TenYearCHD'] == 1]
heart_disease
```

```python
#Group by age column and count the rows of TenYearCHD column
grp_age = heart_disease.groupby('age')['TenYearCHD'].count().reset_index()
grp_age
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years
plt.figure(figsize=(20,8))

plt.scatter(grp_age['age'], grp_age['TenYearCHD'],color='green', label='Ten Year CHD of age groups',alpha=0.5, s = grp_age['TenYearCHD']*10)


plt.legend()
plt.xlabel("Age", size=14)
plt.ylabel("Ten Year CHD", size=14)
plt.title("Graph Showing the chance of people having coronary heart disease in 10 years", size=20)
```

Conslusion - 

```python
Understanding from this is that the bigger the bubble is, the higher the chance is of the person having coronary heart disease in 10 years
```
